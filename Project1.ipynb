```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(rc={"figure.figsize":(20,12)}) # change the figure size to width=20, height=12
sns.lineplot(data=ps4Data_df, x='Year', y='Global', hue='Publisher')
```

```python
graph1 = sns.scatterplot(data=ps4Data_df, x='Year', y='Global', hue='Genre', palette='Accent')
graph1.set_xlabel('Year', fontsize=25)
graph1.set_ylabel('Global Sales #', fontsize=25)
graph1.legend(title='Genres', loc="upper right", frameon=True, fontsize=15)
```

```python
graph2 = sns.scatterplot(data=xboxData_df, x='Year', y='Global', hue='Genre', palette='Accent')
graph2.set_xlabel('Year', fontsize=25)
graph2.set_ylabel('Global Sales #', fontsize=25)
graph2.legend(title='Genres', loc="upper right", frameon=True, fontsize=15)
```

```python
graph3 = sns.scatterplot(data=gamingData_df, x='Year_of_Release', y='Global_Sales', hue='Genre', palette='Accent')
graph3.set_xlabel('Year', fontsize=25)
graph3.set_ylabel('Global Sales #', fontsize=25)
graph3.legend(title='Genres', loc="upper right", frameon=True, fontsize=15)
```

```python
graph4 = sns.barplot(data=ps4Data_df, x='Year', y='Global', hue='Genre', palette='Accent')
graph4.legend(title='Genres', loc="upper right", frameon=True, fontsize=15)
```
